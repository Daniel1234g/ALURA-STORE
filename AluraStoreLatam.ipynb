### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular las ventas
Venta_tienda = tienda['Precio'].sum()
Venta_tienda2 = tienda2['Precio'].sum()
Venta_tienda3 = tienda3['Precio'].sum()
Venta_tienda4 = tienda4['Precio'].sum()

# Crear lista con tuplas (nombre_tienda, total_ventas)
ventas = [
    ("Tienda 1", Venta_tienda),
    ("Tienda 2", Venta_tienda2),
    ("Tienda 3", Venta_tienda3),
    ("Tienda 4", Venta_tienda4)
]

# Imprimir ventas individuales
for nombre, total in ventas:
    print(f"{nombre}: ${total:.2f}")

# Ordenar de menor a mayor por total de ventas
ventas.sort(key=lambda x: x[1])

print("\nVentas de menor a mayor:")
for nombre, total in ventas:
    print(f"{nombre}: ${total:.2f}")
```

# 2. Ventas por categoría

```python
# Tienda 1
print("Categorías más vendidas - Tienda 1")
print(tienda['Categoría del Producto'].value_counts())
print("\n")

# Tienda 2
print("Categorías más vendidas - Tienda 2")
print(tienda2['Categoría del Producto'].value_counts())
print("\n")

# Tienda 3
print("Categorías más vendidas - Tienda 3")
print(tienda3['Categoría del Producto'].value_counts())
print("\n")

# Tienda 4
print("Categorías más vendidas - Tienda 4")
print(tienda4['Categoría del Producto'].value_counts())
print("\n")
```

# 3. Calificación promedio de la tienda


```python
calif1 = tienda['Calificación'].mean()
calif2 = tienda2['Calificación'].mean()
calif3 = tienda3['Calificación'].mean()
calif4 = tienda4['Calificación'].mean()

print(f"Tienda 1: {calif1:.2f}")
print(f"Tienda 2: {calif2:.2f}")
print(f"Tienda 3: {calif3:.2f}")
print(f"Tienda 4: {calif4:.2f}")

calificaciones = [
    ("Tienda 1", calif1),
    ("Tienda 2", calif2),
    ("Tienda 3", calif3),
    ("Tienda 4", calif4)
]

calificaciones.sort(key=lambda x: x[1], reverse=True)

print("\nTiendas ordenadas por calificación promedio:")
for nombre, calif in calificaciones:
    print(f"{nombre}: {calif:.2f}")
```

# 5. Envío promedio por tienda

```python
envio1 = tienda['Costo de envío'].mean()
envio2 = tienda2['Costo de envío'].mean()
envio3 = tienda3['Costo de envío'].mean()
envio4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio - Tienda 1: ${envio1:.2f}")
print(f"Costo de envío promedio - Tienda 2: ${envio2:.2f}")
print(f"Costo de envío promedio - Tienda 3: ${envio3:.2f}")
print(f"Costo de envío promedio - Tienda 4: ${envio4:.2f}")

import matplotlib.pyplot as plt

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos = [envio1, envio2, envio3, envio4]

plt.figure(figsize=(8,4))
plt.bar(nombres, costos, color='lightgreen')
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo Promedio ($)")
plt.ylim(0, max(costos) + 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
