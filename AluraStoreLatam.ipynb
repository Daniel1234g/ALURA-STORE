### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
# Calcular las ventas
Venta_tienda = tienda['Precio'].sum()
Venta_tienda2 = tienda2['Precio'].sum()
Venta_tienda3 = tienda3['Precio'].sum()
Venta_tienda4 = tienda4['Precio'].sum()

# Crear lista con tuplas (nombre_tienda, total_ventas)
ventas = [
    ("Tienda 1", Venta_tienda),
    ("Tienda 2", Venta_tienda2),
    ("Tienda 3", Venta_tienda3),
    ("Tienda 4", Venta_tienda4)
]

# Imprimir ventas individuales
for nombre, total in ventas:
    print(f"{nombre}: ${total:.2f}")

# Ordenar de menor a mayor por total de ventas
ventas.sort(key=lambda x: x[1])

print("\nVentas de menor a mayor:")
for nombre, total in ventas:
    print(f"{nombre}: ${total:.2f}")
```

# 2. Ventas por categor√≠a

```python
# Tienda 1
print("Categor√≠as m√°s vendidas - Tienda 1")
print(tienda['Categor√≠a del Producto'].value_counts())
print("\n")

# Tienda 2
print("Categor√≠as m√°s vendidas - Tienda 2")
print(tienda2['Categor√≠a del Producto'].value_counts())
print("\n")

# Tienda 3
print("Categor√≠as m√°s vendidas - Tienda 3")
print(tienda3['Categor√≠a del Producto'].value_counts())
print("\n")

# Tienda 4
print("Categor√≠as m√°s vendidas - Tienda 4")
print(tienda4['Categor√≠a del Producto'].value_counts())
print("\n")
```

# 3. Calificaci√≥n promedio de la tienda


```python
calif1 = tienda['Calificaci√≥n'].mean()
calif2 = tienda2['Calificaci√≥n'].mean()
calif3 = tienda3['Calificaci√≥n'].mean()
calif4 = tienda4['Calificaci√≥n'].mean()

print(f"Tienda 1: {calif1:.2f}")
print(f"Tienda 2: {calif2:.2f}")
print(f"Tienda 3: {calif3:.2f}")
print(f"Tienda 4: {calif4:.2f}")

calificaciones = [
    ("Tienda 1", calif1),
    ("Tienda 2", calif2),
    ("Tienda 3", calif3),
    ("Tienda 4", calif4)
]

calificaciones.sort(key=lambda x: x[1], reverse=True)

print("\nTiendas ordenadas por calificaci√≥n promedio:")
for nombre, calif in calificaciones:
    print(f"{nombre}: {calif:.2f}")
```

```python
# Tienda 1
print("üîù Producto m√°s vendido - Tienda 1:")
print(tienda['Producto'].value_counts().head(1))
print("üîª Producto menos vendido - Tienda 1:")
print(tienda4['Producto'].value_counts().tail(1))

# Tienda 2
print("üîù Producto m√°s vendido - Tienda 2:")
print(tienda2['Producto'].value_counts().head(1))
print("üîª Producto menos vendido - Tienda 2:")
print(tienda2['Producto'].value_counts().tail(1))

# Tienda 3
print("üîù Producto m√°s vendido - Tienda 3:")
print(tienda3['Producto'].value_counts().head(1))
print("üîª Producto menos vendido - Tienda 3:")
print(tienda3['Producto'].value_counts().tail(1))

# Tienda 4
print("üîù Producto m√°s vendido - Tienda 4:")
print(tienda4['Producto'].value_counts().head(1))
print("üîª Producto menos vendido - Tienda 4:")
print(tienda4['Producto'].value_counts().tail(1))

import matplotlib.pyplot as plt

# Top 5 productos Tienda 1
top5_t1 = tienda['Producto'].value_counts().head(5)

plt.figure(figsize=(8,4))
top5_t1.plot(kind='bar', color='skyblue')
plt.title("Top 5 Productos M√°s Vendidos - Tienda 1")
plt.ylabel("Cantidad vendida")
plt.xlabel("Producto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Top 5 productos Tienda 2
top5_t1 = tienda2['Producto'].value_counts().head(5)

plt.figure(figsize=(8,4))
top5_t1.plot(kind='bar', color='skyblue')
plt.title("Top 5 Productos M√°s Vendidos - Tienda 2")
plt.ylabel("Cantidad vendida")
plt.xlabel("Producto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Top 5 productos Tienda 1
top5_t1 = tienda3['Producto'].value_counts().head(5)

plt.figure(figsize=(8,4))
top5_t1.plot(kind='bar', color='skyblue')
plt.title("Top 5 Productos M√°s Vendidos - Tienda 3")
plt.ylabel("Cantidad vendida")
plt.xlabel("Producto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Top 5 productos Tienda 1
top5_t1 = tienda4['Producto'].value_counts().head(5)

plt.figure(figsize=(8,4))
top5_t1.plot(kind='bar', color='skyblue')
plt.title("Top 5 Productos M√°s Vendidos - Tienda 4")
plt.ylabel("Cantidad vendida")
plt.xlabel("Producto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 5. Env√≠o promedio por tienda

```python
envio1 = tienda['Costo de env√≠o'].mean()
envio2 = tienda2['Costo de env√≠o'].mean()
envio3 = tienda3['Costo de env√≠o'].mean()
envio4 = tienda4['Costo de env√≠o'].mean()

print(f"Costo de env√≠o promedio - Tienda 1: ${envio1:.2f}")
print(f"Costo de env√≠o promedio - Tienda 2: ${envio2:.2f}")
print(f"Costo de env√≠o promedio - Tienda 3: ${envio3:.2f}")
print(f"Costo de env√≠o promedio - Tienda 4: ${envio4:.2f}")

import matplotlib.pyplot as plt

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos = [envio1, envio2, envio3, envio4]

plt.figure(figsize=(8,4))
plt.bar(nombres, costos, color='lightgreen')
plt.title("Costo de Env√≠o Promedio por Tienda")
plt.ylabel("Costo Promedio ($)")
plt.ylim(0, max(costos) + 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
