### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

# Calcular las ventas
Venta_tienda = tienda['Precio'].sum()
Venta_tienda2 = tienda2['Precio'].sum()
Venta_tienda3 = tienda3['Precio'].sum()
Venta_tienda4 = tienda4['Precio'].sum()

# Crear lista con tuplas (nombre_tienda, total_ventas)
ventas = [
    ("Tienda 1", Venta_tienda),
    ("Tienda 2", Venta_tienda2),
    ("Tienda 3", Venta_tienda3),
    ("Tienda 4", Venta_tienda4)
]

# Imprimir ventas individuales
for nombre, total in ventas:
    print(f"{nombre}: ${total:.2f}")

# Ordenar de menor a mayor por total de ventas
ventas.sort(key=lambda x: x[1])

print("\nVentas de menor a mayor:")
for nombre, total in ventas:
    print(f"{nombre}: ${total:.2f}")

# Preparar datos para gráfico de líneas
nombres = [nombre for nombre, _ in ventas]
totales = [total for _, total in ventas]

# Crear gráfico de líneas
plt.figure(figsize=(8, 5))
plt.plot(nombres, totales, marker='o', linestyle='-', color='b')
plt.title("Ventas Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Total de Ventas ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
import matplotlib.pyplot as plt
import pandas as pd

# Supongamos que ya tienes tus DataFrames: tienda, tienda2, tienda3, tienda4

# Calcular la calificación promedio de cada tienda
calif1 = tienda['Calificación'].mean()
calif2 = tienda2['Calificación'].mean()
calif3 = tienda3['Calificación'].mean()
calif4 = tienda4['Calificación'].mean()

# Imprimir calificaciones
print(f"Tienda 1: {calif1:.2f}")
print(f"Tienda 2: {calif2:.2f}")
print(f"Tienda 3: {calif3:.2f}")
print(f"Tienda 4: {calif4:.2f}")

# Guardar las calificaciones en una lista
calificaciones = [
    ("Tienda 1", calif1),
    ("Tienda 2", calif2),
    ("Tienda 3", calif3),
    ("Tienda 4", calif4)
]

# Ordenar por calificación promedio (de mayor a menor)
calificaciones.sort(key=lambda x: x[1], reverse=True)

# Imprimir ordenado
print("\nTiendas ordenadas por calificación promedio:")
for nombre, calif in calificaciones:
    print(f"{nombre}: {calif:.2f}")

# Crear gráfico de pastel
nombres = [nombre for nombre, _ in calificaciones]
valores = [calif for _, calif in calificaciones]

plt.figure(figsize=(6, 6))
plt.pie(valores, labels=nombres, autopct='%1.1f%%', startangle=90)
plt.title("Distribución de Calificaciones Promedio por Tienda")
plt.axis('equal')  # Para que el gráfico sea circular
plt.show()
```

# 5. Envío promedio por tienda

```python
envio1 = tienda['Costo de envío'].mean()
envio2 = tienda2['Costo de envío'].mean()
envio3 = tienda3['Costo de envío'].mean()
envio4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio - Tienda 1: ${envio1:.2f}")
print(f"Costo de envío promedio - Tienda 2: ${envio2:.2f}")
print(f"Costo de envío promedio - Tienda 3: ${envio3:.2f}")
print(f"Costo de envío promedio - Tienda 4: ${envio4:.2f}")

import matplotlib.pyplot as plt

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos = [envio1, envio2, envio3, envio4]

plt.figure(figsize=(8,4))
plt.bar(nombres, costos, color='lightgreen')
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo Promedio ($)")
plt.ylim(0, max(costos) + 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
